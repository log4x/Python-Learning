```python
import math

def magnitude(vector): 
    return math.sqrt(sum(pow(element, 2) for element in vector))
```

```python
v = np.array([0, 1, 2, 3, 4])
print('Vector:', v)
```

```python

print(magnitude(v))
```

```python
# Method 2
v = np.array([3.0, -4.0, 12.0])

print(v)
print()


print("Euclidean norm (||v||):", la.norm(v))
print("1-norm:", la.norm(v, 1))
print("Infinity norm:", la.norm(v, np.inf))
```

```python
# EIGENVALUES & EIGENVECTORS

A_eig = np.array([
    [4.0, 2.0],
    [1.0, 3.0]
])
```

```python
eigvals, eigvecs = la.eig(A_eig)
```

```python
print("\nEigenvalues:", eigvals)
```

```python

print("\nRight Eigenvectors (columns):\n", eigvecs)
```

```python
print("\n-- Verification of A·v = λ·v --")
for i, λ in enumerate(eigvals):
    v = eigvecs[:, i]
    print(f"Eigenpair {i}: ", np.allclose(A_eig @ v, λ * v))
```

```python
eigvals_only = la.eigvals(A_eig)
print("eigvals(A_eig):", eigvals_only)
```

```python
detA = la.det(A_eig)
print("det(A_eig):", detA)
```

```python
print("\n==================== INVERSE OF MATRIX ====================\n")

if abs(detA) < 1e-12:
    print("Matrix is singular. No inverse exists.")
else:
    Ainv = la.inv(A_eig)
    print("Inverse of A_eig:\n", Ainv)

    print("\nA @ A^-1 (should be identity):\n", np.round(A_eig @ Ainv, 10))
```

```python
A = np.array([10, 20, 30, 40])
B = np.array([1, 2, 3, 4])

print("A:", A)
print("B:", B)
```

```python
# ---- Arithmetic operations ----
print("\nArithmetic:")
print("A + B    ->", A + B)       # element-wise addition
print("A - B    ->", A - B)       # element-wise subtraction
print("A * B    ->", A * B)       # element-wise multiplication
print("A / B    ->", A / B)       # element-wise division (float)
print("A // B   ->", A // B)      # floor division
print("A % B    ->", A % B)       # remainder / modulus
print("A ** B   ->", A ** B)      # element-wise exponentiation
```

```python
C = A.copy()
C += 5                            # in-place addition
print("\nC after C += 5:", C)
```

```python
# ---- Comparisons ----
print("\nComparisons:")
print("A == B ->", A == B)        # equal comparison
print("A != B ->", A != B)        # not equal comparison
print("A > B  ->", A > B)         # greater-than
print("A <= B ->", A <= B)        # less-than-or-equal
```

```python
bool1 = np.array([True, False, True, False])
bool2 = np.array([True, True, False, False])
```

```python
print("\nLogical operations:")
print("logical_and ->", np.logical_and(bool1, bool2))   # AND
print("logical_or  ->", np.logical_or(bool1, bool2))    # OR
print("logical_xor ->", np.logical_xor(bool1, bool2))   # XOR
print("logical_not(bool1) ->", np.logical_not(bool1))   # NOT
```

```python
# ---- Bitwise operations ----
X = np.array([0b0011, 0b0101])    # integers 3,5
Y = np.array([0b0101, 0b0010])    # integers 5,2
```

```python
print("\nBitwise operations:")
print("X & Y  ->", np.bitwise_and(X, Y))     # bitwise AND
print("X | Y  ->", np.bitwise_or(X, Y))      # bitwise OR
print("X ^ Y  ->", np.bitwise_xor(X, Y))     # bitwise XOR
print("~X     ->", np.bitwise_not(X))        # bitwise NOT
print("X << 1 ->", np.left_shift(X, 1))      # left shift
print("X >> 1 ->", np.right_shift(X, 1))     # right shift
```

```python
# ---- Rounding functions ----
mixed = np.array([1.2, 2.7, -1.3, -2.8])
print("\nRounding:")
print("np.floor(mixed) ->", np.floor(mixed)) # round down
print("np.ceil(mixed)  ->", np.ceil(mixed))  # round up
print("np.round(mixed) ->", np.round(mixed)) # round to nearest
print("np.trunc(mixed) ->", np.trunc(mixed)) # truncate fraction
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("\nAggregations:")
print("np.sum(arr) ->", np.sum(arr))         # total sum
print("np.sum(arr, axis=0) ->", np.sum(arr, axis=0))  # column-wise sum
print("np.mean(arr) ->", np.mean(arr))       # mean
print("np.std(arr)  ->", np.std(arr))        # standard deviation
print("np.argmin(arr) ->", np.argmin(arr))   # returns index of min
print("np.argmax(arr) ->", np.argmax(arr))   # returns index of max
```

```python
s1 = np.array(["hello", "world", "NumPy"])
s2 = np.array(["!", "!!!", "rocks"])
```

```python
# ---- Concatenation ----
print("\nConcatenation:")
print(np.char.add(s1, s2))       # element-wise string add
print(np.char.multiply(np.array(["ha","ho"]), 3))          # repeat string
```

```python
# ---- Case conversions & strip ----

print(np.char.upper(s1))           # uppercase
print(np.char.lower(s1))           # lowercase
print(np.char.capitalize(np.array(["numpy array"])))       # capitalize
print(np.char.strip(np.array(["  a "])) )                  # remove whitespace
```

```python
# ---- Join, split, replace ----

print(np.char.join(' ', np.array(["a","b","c"])))          # join chars
print()

print(np.char.split(np.array(["one two"])))                # split string
print()

print(np.char.replace(np.array(["foo"]), "o", "0"))        # replace substring
```

```python
arr = np.char.array(["one two"])

chars = list(arr[0])

chars_no_space = [c for c in chars if c != ' ']
print(chars_no_space)

```

```python
# ---- Padding & formatting ----
fmt = np.array(["7", "42", "123456"])

print(np.char.zfill(fmt, 5))   # pad with zeros
print(np.char.center(s1, 11, '*'))  # center
print(np.char.ljust(s1, 10))   # left-justify
print(np.char.rjust(s1, 10))   # right-justify
```

```python
# ---- Searching ----

print(s1)
print()

print(np.char.startswith(s1, "h"))  # starts with
print(np.char.endswith(np.char.lower(s1), "py"))                     # ends with
print(np.char.find(s1, "l"))             # find 1st index where alphabet matches
```

```python
# ---- Convert String to number & number to string ----
nums = np.array([1,2,3])


strs = np.char.mod('%d', nums)        # format numbers to strings
print(strs)
print()

back = strs.astype(int)
print(back)   # convert back
```
