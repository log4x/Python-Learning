```python
def gaussian_filter(kernel_size, sigma=1, muu=0):
    # Initializing value of x, y as grid of kernel size in the range of kernel size
    x, y = np.meshgrid(np.linspace(-1, 1, kernel_size),
                       np.linspace(-1, 1, kernel_size))
    dst = np.sqrt(x**2 + y**2)

    # Normal part of the Gaussian function
    normal = 1 / (2 * np.pi * sigma**2)

    # Calculating Gaussian filter
    gauss = np.exp(-((dst - muu)**2 / (2.0 * sigma**2))) * normal

    return gauss  # Return the calculated Gaussian filter


# Example usage:
kernel_size = 5
gaussian = gaussian_filter(kernel_size)
print("Gaussian filter of {} X {}:".format(kernel_size, kernel_size))
print(gaussian)
```

### Combining Arrays

numpy.hstack() function stacks arrays in sequence horizontally (column-wise). It joins arrays along their second axis for 2D arrays or flattens and joins them for 1D arrays. This is useful for combining arrays side by side.

```python
a = np.array([[1, 2], [3, 4]])
print(a)

b = np.array([[5, 6], [7, 8]])
print(b)

res = np.hstack((a, b))
print(res)
```

```python
a = np.array([[1, 2], [3, 4]])
print(a)

b = np.array([[5, 6], [7, 8]])
print(b)

res = np.vstack((a, b))
print(res)
```

dstack stacks them depth-wise to form a 3D array.

```python
a = np.array([[1, 2], [3, 4]])
print(a)

b = np.array([[5, 6], [7, 8]])
print(b)

res = np.dstack((a, b))
print(res)
```

The nditer function in NumPy is an iterator that allows you to efficiently loop through arrays, even if they have different shapes. When you give it two arrays (like 1-D and 2-D), it pairs up their elements side by side and lets you process them together.

```python
num1 = np.array([1, 2,])
num2 = np.array([[10, 20],
                 [30, 40]])
                   
for a, b in np.nditer([num1, num2]):
    print(a, ":", b , "=", a*b)
```

np.ma.concatenate() method joins two or more masked arrays along an existing axis. It works similarly to np.concatenate() but it is specifically designed for masked arrays which may contain missing or invalid data that is "masked" i.e marked to be ignored in computations.

```python
a = np.ma.array([1, 2, 3], mask=[0, 1, 0])
b = np.ma.array([4, 5, 6], mask=[0, 0, 1])

res = np.ma.concatenate([a, b])
print(res)
```

Finding unique rows means removing duplicate rows from a 2D array so that each row appears only once. For example, given [[1, 2], [3, 4], [1, 2], [5, 6]], the unique rows would be [[1, 2], [3, 4], [5, 6]]. Let’s explore efficient ways to achieve this using NumPy.

```python
a = np.array([1, 2, 2, 3, 4, 4, 4])
res = np.unique(a)
print(res)
```

```python
a = np.array([[1, 2], [3, 4], [1, 2], [5, 6]])
res = np.unique(a, axis=0)
print(res)
```

```python
a = np.array([0, 0, 3, 4, 0, 5, 0, 0])
res = np.trim_zeros(a)
print(res)
```

### Combined Code

```python
# combined_array_ops.py
# Demonstrates copy/view, append, swap columns, newaxis, stacking/joining, splitting,
# comparisons, union/unique/unique-rows, masked concat, and trim_zeros — all at top-level.

import numpy as np

print("\n=== 1) copy vs view ===")
a = np.array([1, 2, 3, 4])
v = a.view()           # view - shared data
c = a.copy()           # deep copy - separate data
print("original a:", a)
v[0] = 99
print("after v[0]=99 -> a (changed because view shares memory):", a)
c[1] = 88
print("after c[1]=88 -> a (unchanged):", a, "c:", c)

print("\n=== 2) copy array into another array (np.copy & slicing) ===")
src = np.arange(6)
dst = np.empty_like(src)
np.copyto(dst, src)    # copyto
print("src:", src)
print("dst after np.copyto:", dst)
# slicing copy example
dst2 = src[:]          # creates a copy for 1-D? actually makes a view for arrays, use .copy() if you want deep copy
dst2 = src.copy()
print("dst2 (src.copy()):", dst2)

print("\n=== 3) append values at end (np.append) ===")
arr = np.array([1, 2, 3])
arr_app = np.append(arr, [4,5])
print("arr:", arr, " after append ->", arr_app)
# for 2D use axis
mat = np.array([[1,2],[3,4]])
mat_app = np.append(mat, [[5,6]], axis=0)
print("mat:\n", mat, "\nmat after append row:\n", mat_app)

print("\n=== 4) swap columns of a given numpy array ===")
M = np.array([[10,20,30],
              [40,50,60]])
print("Before swap:\n", M)
# swap col 0 and 2
M[:, [0,2]] = M[:, [2,0]]
print("After swapping col0 and col2:\n", M)

print("\n=== 5) insert a new axis within a numpy array (np.newaxis / expand_dims) ===")
r = np.array([1,2,3])
print("r shape:", r.shape)
r_row = r[np.newaxis, :]        # shape (1,3)
r_col = r[:, np.newaxis]        # shape (3,1)
print("r_row shape:", r_row.shape, " ->\n", r_row)
print("r_col shape:", r_col.shape, " ->\n", r_col)
# using expand_dims
r_exp = np.expand_dims(r, axis=0)
print("expand_dims axis=0 shape:", r_exp.shape)

print("\n=== 6) np.hstack ===")
a = np.array([1,2,3])
b = np.array([4,5,6])
hs = np.hstack((a, b))
print("hstack 1D:", hs)
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("hstack 2D:\n", np.hstack((A,B)))

print("\n=== 7) np.vstack ===")
print("vstack 1D -> stacks as rows:\n", np.vstack((a,b)))
print("vstack 2D:\n", np.vstack((A,B)))

print("\n=== 8) joining arrays (np.concatenate) ===")
print("concatenate along axis=0 (rows):\n", np.concatenate((A,B), axis=0))
print("concatenate along axis=1 (cols):\n", np.concatenate((A,B), axis=1))

print("\n=== 9) combining a one- and a two-dimensional array ===")
one = np.array([9,9])
two = np.array([[1,2],[3,4]])
# to combine, align dimensions (make one into row or column)
one_row = one[np.newaxis, :]      # shape (1,2)
combined_vert = np.vstack((two, one_row))  # now shapes match
print("two:\n", two, "\none_row:\n", one_row, "\ncombined_vert:\n", combined_vert)

print("\n=== 10) np.ma.concatenate (masked-array concatenate) ===")
import numpy.ma as ma
m1 = ma.array([1,2,3])
m2 = ma.array([4,5], mask=[0,1])   # second element masked
mcat = ma.concatenate((m1, m2))
print("m1:", m1, "m2:", m2, "concatenated masked array:", mcat)

print("\n=== 11) np.dstack (depth stack) ===")
x = np.array([[1,2],[3,4]])
y = np.array([[5,6],[7,8]])
print("x:\n", x, "\ny:\n", y)
d = np.dstack((x,y))   # stacks along third axis -> shape (2,2,2)
print("dstack -> shape", d.shape, "\n", d)

print("\n=== 12) splitting arrays in numpy (hsplit, vsplit, array_split) ===")
Z = np.arange(12).reshape(3,4)
print("Z:\n", Z)
hleft, hright = np.hsplit(Z, [2])   # split after column 2
print("hsplit into [0:2] and [2:]:\n", hleft, "\n", hright)
vtop, vbottom = np.vsplit(Z, [1])  # split after row 1
print("vsplit into [0:1] and [1:]:\n", vtop, "\n", vbottom)
parts = np.array_split(Z, 3, axis=1)  # flexible split
print("array_split into 3 parts along columns -> lengths:", [p.shape for p in parts])

print("\n=== 13) how to compare two numpy arrays ===")
A1 = np.array([1,2,3])
A2 = np.array([1,2,4])
print("element-wise comparison (A1==A2):", A1 == A2)
print("all elements equal (np.array_equal):", np.array_equal(A1, A2))
print("positions different (np.where):", np.where(A1 != A2))

print("\n=== 14) union of two numpy arrays (np.union1d) ===")
u1 = np.array([1,2,3,3])
u2 = np.array([3,4,5])
print("union1d(u1,u2):", np.union1d(u1, u2))

print("\n=== 15) find unique rows in a numpy array ===")
R = np.array([[1,2],
              [1,2],
              [3,4],
              [3,4],
              [5,6]])
uniq_rows = np.unique(R, axis=0)
print("R:\n", R)
print("unique rows:\n", uniq_rows)

print("\n=== 16) np.unique examples ===")
vals = np.array([1,2,2,3,3,3])
uniq, counts = np.unique(vals, return_counts=True)
print("values:", vals, "unique:", uniq, "counts:", counts)
# unique with return_index, return_inverse
uniq2, idx = np.unique([3,1,2,1,3], return_index=True)
print("unique:", uniq2, "first-occurrence indices:", idx)

print("\n=== 17) np.trim_zeros ===")
z = np.array([0,0,1,2,0,3,0,0])
print("original:", z)
print("trim_zeros (both ends):", np.trim_zeros(z, 'fb'))   # 'f' front, 'b' back, 'fb' both
print("trim_zeros front only:", np.trim_zeros(z, 'f'))
print("trim_zeros back only:", np.trim_zeros(z, 'b'))

print("\n=== Done ===")
```

```python
# combined_linear_algebra_ops_clean.py
# Linear Algebra Operations: QR, Norm, Eigenvalues, Eigvals, Determinant, Inverse
# All code is top-level, simple & spaced for readability.

import numpy as np
from numpy import linalg as LA



# ============================================================
# 1) QR FACTORIZATION
# ============================================================

print("\n==================== QR FACTORIZATION ====================\n")

A_qr = np.array([
    [12.0, -51.0,   4.0],
    [ 6.0, 167.0, -68.0],
    [-4.0,  24.0, -41.0]
])

print("Matrix A_qr:\n", A_qr)

Q, R = LA.qr(A_qr)

print("\nQ Matrix:\n", Q)
print("\nR Matrix:\n", R)

print("\nReconstructed (Q @ R):\n", np.round(Q @ R, 10))
print("Is Q orthonormal?:", np.allclose(Q.T @ Q, np.eye(Q.shape[1])))



# ============================================================
# 2) VECTOR MAGNITUDE / NORM
# ============================================================

print("\n==================== VECTOR MAGNITUDE ====================\n")

v = np.array([3.0, -4.0, 12.0])

print("Vector v:", v)
print("Euclidean norm (||v||):", LA.norm(v))
print("1-norm:", LA.norm(v, 1))
print("Infinity norm:", LA.norm(v, np.inf))



# ============================================================
# 3) EIGENVALUES & EIGENVECTORS (np.linalg.eig)
# ============================================================

print("\n==================== EIGENVALUES & EIGENVECTORS ====================\n")

A_eig = np.array([
    [4.0, 2.0],
    [1.0, 3.0]
])

print("Matrix A_eig:\n", A_eig)

eigvals, eigvecs = LA.eig(A_eig)

print("\nEigenvalues:", eigvals)
print("\nRight Eigenvectors (columns):\n", eigvecs)

print("\n-- Verification of A·v = λ·v --")
for i, λ in enumerate(eigvals):
    v = eigvecs[:, i]
    print(f"Eigenpair {i}: ", np.allclose(A_eig @ v, λ * v))



# ============================================================
# 4) EIGENVALUES ONLY (np.linalg.eigvals)
# ============================================================

print("\n==================== EIGENVALUES ONLY ====================\n")

eigvals_only = LA.eigvals(A_eig)
print("eigvals(A_eig):", eigvals_only)



# ============================================================
# 5) DETERMINANT
# ============================================================

print("\n==================== DETERMINANT ====================\n")

detA = LA.det(A_eig)
print("det(A_eig):", detA)



# ============================================================
# 6) INVERSE
# ============================================================

print("\n==================== INVERSE OF MATRIX ====================\n")

if abs(detA) < 1e-12:
    print("Matrix is singular. No inverse exists.")
else:
    Ainv = LA.inv(A_eig)
    print("Inverse of A_eig:\n", Ainv)

    print("\nA @ A^-1 (should be identity):\n", np.round(A_eig @ Ainv, 10))



# ============================================================
# 7) EXTRA: 3x3 EXAMPLE
# ============================================================

print("\n==================== EXTRA: 3x3 MATRIX EXAMPLE ====================\n")

B = np.array([
    [1.0, 2.0, 3.0],
    [0.0, 1.0, 4.0],
    [5.0, 6.0, 0.0]
])

print("Matrix B:\n", B)

print("\nEigenvalues of B:", LA.eigvals(B))
print("det(B):", LA.det(B))

try:
    Binv = LA.inv(B)
    print("\nInverse of B:\n", Binv)
except LA.LinAlgError:
    print("\nB is singular — no inverse exists.")



print("\n==================== DONE ====================\n")
```
